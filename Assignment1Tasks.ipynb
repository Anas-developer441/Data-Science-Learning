```python
# Task 6: Floor Division (The "Baker's" Problem)
#You have 22 cookies and boxes that hold 5 cookies each. How many full boxes can you ship?
total_cookies=22
#so there are boxes who hold 5 cookies so
cookies_in_box=5
# im gonna use divide method
#initilizing variable and calculating 
Boxes= total_cookies // cookies_in_box
print("THE TOTAL BOXES EACH  CONTAINS 5 COOKIES ARE: ",Boxes)
#and we did use // this because we dont want answer in points 
```

```python
#Task 7: Compound Interest
#Calculate the final balance of an investment. Use principal = 1000, rate = 0.05, and time = 3.

#INITILIZE VARIABLES 
#formula is A= P X (1 + r) ^t
principal = 1000
rate=0.05
time=3
#put inside the formula and store the value in variable final_balance
final_balance= principal * ( 1 + rate)**time
print("FINAL BALANCE IS: ",final_balance)

```
