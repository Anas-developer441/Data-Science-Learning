```python
# COUNTING TOTAL WEATHERS AND SEPARATING THE SAME AND PRINTING THEM ALONG THEIR LogID

import csv

# open csv file using utf-8 encoding
file = open('solar_grid.csv', encoding='utf-8')

# convert csv reader into list
solar_Data = list(csv.reader(file))

# seperate Header and Row
header = solar_Data[0]
rows = solar_Data[1:]

# find the index of the weather column
w = header.index("Weather")

# creating empty list so that different weather can be stored inside it
WEATHERS = []

# creating different lists to store same weather values in different lists
sunny = []
cloudy = []
partlyCloudy = []
Rainy = []

# it will count how many total weather records exist in file
weather_count = 0

print("Header", header)

# loop through each row
for i, data in enumerate(rows, start=1):

    # now i want to create a weather variable
    weather = data[w]

    # counting total weather records
    weather_count += 1

    # storing weather types in WEATHERS list
    WEATHERS.append(weather)

    # if else to append same weather group in each list
    # storing row number and row data as tuple

    if weather == "Sunny":
        sunny.append((i, data))

    elif weather == "Partly Cloudy":
        partlyCloudy.append((i, data))

    elif weather == "Rainy":
        Rainy.append((i, data))

    elif weather == "Cloudy":
        cloudy.append((i, data))


# printing total weather records
print("\nTotal weathers =", weather_count)

# printing all weather types found in dataset
print("\nAll Weather Types:")
for j, dataa in enumerate(WEATHERS, start=1):
    print(j, dataa)


# printing sunny weather rows
print("\nSunny Days:")
for a, sunn in enumerate(sunny, start=1):
    print(a, sunn)

# printing partly cloudy weather rows
print("\nPartly Cloudy Days:")
for b, parCloud in enumerate(partlyCloudy, start=1):
    print(b, parCloud)

# printing cloudy weather rows
print("\nCloudy Days:")
for c, cloud in enumerate(cloudy, start=1):
    print(c, cloud)

# printing rainy weather rows
print("\nRainy Days:")
for d, Rain in enumerate(Rainy, start=1):
    print(d, Rain)
```

```python
#                                    DICTIONARY

# also check average of VOLTAGES CURRENT_AMPS AND HOUR 
import csv
file=open('solar_grid.csv',encoding='utf-8')
solarlist=list(csv.reader(file))

#seperating header and rows
header=solarlist[0]
rows=solarlist[1:]

#write the index of those variables u want to work on 
H=header.index("Hour")
V=header.index("Voltage")
C_Amp=header.index("Current_Amps")
W=header.index("Weather")


#                               task
#to seperate  different weathers and arrange same weathers together in their list 



#                            use dictionary 
dictionaryWeather={"SUNNY":[],"PARTLYCLOUDY":[],"CLOUDY":[],"RAINY":[]} #four lists
AVERAGE={"HOUR":[],"VOLTAGE":[],"CURRENTAMPERES":[]}
total_weather_days=0
for i,row in enumerate(rows,start=1):
    Hour=int(row[H])
    Voltage=float(row[V])
    Current_Amp=float(row[C_Amp])
    weather=row[w]
    #STORE VALUES FOR AVERAGE CALCULATION 
    AVERAGE["HOUR"].append(Hour)
    AVERAGE["VOLTAGE"].append(Voltage)
    AVERAGE["CURRENTAMPERES"].append(Current_Amp)

    
    #now count how mamy weathers are in the list 
    total_weather_days+=1
    if weather=="Sunny":
        value="SUNNY"
    elif weather == "Partly Cloudy":
        value="PARTLYCLOUDY"
    elif weather == "Cloudy":
        value="CLOUDY"
    elif weather == "Rainy":
        value="RAINY"
   #now append those conditions which are true in their dictionary 
#you just have to put value inside the dictionary and write what u want to print an output
    dictionaryWeather[value].append(row)
    

# print results
for key in dictionaryWeather:
    print("\n", key)
    for record in dictionaryWeather[key]:
        print(record)


print("\nTotal Weather in list:",total_weather_days)


H_average=sum(AVERAGE["HOUR"])/len(AVERAGE["HOUR"])
V_average=sum(AVERAGE["VOLTAGE"])/len(AVERAGE["VOLTAGE"])
CA_average=sum(AVERAGE["CURRENTAMPERES"])/len(AVERAGE["CURRENTAMPERES"])


print("\nHOUR AVERAGE:",H_average)
print("VOLTAGE AVERAGE:",V_average)
print("CURRENT AMPERE AVERAGE:",CA_average)
```
