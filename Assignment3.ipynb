```python
''' Instructions:

Create an empty list called restock_order.
Use a for loop with range(len(inventory)) to go through the list.
After the loop, print: "Total items to order: [count]".'''
# Format: [SKU, Product Name, Current Stock, Minimum Stock Needed]
inventory = [
    ["SKU-101", "Industrial Drill", 45, 50],
    ["SKU-102", "Safety Goggles", 120, 30],
    ["SKU-103", "Steel Bolts", 15, 20],
    ["SKU-104", "Work Gloves", 8, 25],
    ["SKU-105", "Hard Hat", 60, 40]
]


#Create an empty list called restock_order.
restock_order=[]
# use a for loop to go through the list
for a in range(len(inventory)):
  current_stock=inventory[a][2]# a=0 in first run and 2 is where the current stock is present
  minimum_stock=inventory[a][3]# a=0   ->3 where minim stock value is present

    # suing if condition to check which one should be in the list restock_order
  if current_stock<minimum_stock:
        restock_order.append(inventory[a])


print("TOTAl MINIMUM ITEMS COUNT -> ",len(restock_order))
print("TOTAL ITEMS TO ORDER LIST: ",restock_order)
```

```python
regional_sales = [
    # North
    [["John Doe", 12000, 13000, 12500, 14000], ["Jane Smith", 11000, 11500, 12000, 11800]],
    # South
    [["Bob White", 9000, 9500, 9200, 9800], ["Alice Green", 10500, 10800, 11200, 11500]],
    # West
    [["Diana Prince", 15000, 16000, 15500, 17000]]
]

# Step 1: create total_sales variable
total_sales = 0

# Step 2: loop through each region
for region in regional_sales:

    # Step 3: loop through each person in the region
    for person in region:

        # Step 4: add Q1–Q4 sales to total_sales
        for sale in person[1:]:  # 1: means this will include all the values which is present right after index 1
            total_sales += sale

# Step 5: print total sales
print("Total Company Sales:", total_sales)


#basically there are three loops first one will check how many regions are there and it gonna run on that count
# second loop will go inside the region to person and it access person values
# third loop is where we can do inside operation with person list values

```
