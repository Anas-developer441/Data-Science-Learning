```python
#Task 1: Onboarding a New Intern
#create a list called intern with zara , 20,25000,0 and marketing
intern=["zara",20,25000.0,"Marketing"]
basic_info=intern[:3] #means [:3] take index 0,1,2 will add name age salary

# update department which is marketing in the indext so update this to product we will use -1
intern[-1]="Product"
#A bonus is added—append "Onboarding Complete" to the list.
intern.append("Onboarding Complete") #append add this into list

#Later, you realize you need to update the stipend, but you don’t know its index.
#Use .index(25000.0) to find it, then update it to 28000.0.
stipend_index=intern.index(25000.0) #now the index of 25000.0 gonna store in stipend_index
#update the stipend
intern[stipend_index]=28000.0 #now the new value of stipend is updated

#Before archiving, remove "Onboarding Complete" using .remove().
intern.remove("Onboarding Complete")

#Finally, delete the department ("Product") using .pop()
#with the correct index (hint: it’s now the last item).
 #the deptatrment we have is is in index -1 or 3 so we will delete or product department using pop
intern.pop(-1)
#Print the final intern list—it should contain only [name, age, updated_stipend].

print("THE LIST IS: ",intern)


```

```python
#Task 2: Managing a customer Support Ticket

#Create a list ticket with: "Ali", 1045, "High", "Open", 0.0.
ticket = ["Ali",1045,"High","Open",0.0 ]

#Slice to get the core issue info:
#[customer name, issue ID, priority] → store in issue_summary
issue_summary=ticket[:3] # now 4 indexes stored in issue summary

#The ticket is being worked on—update the status ("Open") to "In Progress".
#update the open to In Progress
ticket[3]="In Progress"

#After 2.5 hours, the issue is resolved—append the actual resolution time: 2.5.
#means append 2.5 in list which is resolution time
ticket.append(2.5)

#You need to revise the priority, but aren’t sure where it is.
#Use .index("High") to find its position, then update it to "Medium".
priority_index = ticket.index("High")
ticket[priority_index] = "Medium" # NOW HIGH -> MEDIUM updation

#Now, remove the old resolution time (0.0) using .remove(0.0).
ticket.remove(0.0)


#The system no longer needs the issue ID after resolution—use .pop(1)
#to delete it by index.
ticket.pop(1)


print("The Final Ticket List is: ",ticket)

```
