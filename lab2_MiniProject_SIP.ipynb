```python
# MINI PROJECT
#**********************    SIP(Systematic Investment Plan) CALCULATOR   ********************

#Prompt->i want to create a mini project called sip calculator give me the formulas for that
#not the whole code just the formulas or working so i can code by my own


# input Monthly Investment Amount
Monthly_Inv_Amo=float(input("Enter Monthly Investment Amount="))
Expected_RR=float(input("Enter Expected Rate of Return-Yearly(%)="))
Inv_duration=float(input("Enter Investment Duration[Years]= "))

print("Your Monthly Investment Amount= ",Monthly_Inv_Amo)
print("Your Expected Rate of Return-Yearly(%)=",Expected_RR)
print("Your Investment Duration[Years]= ",Inv_duration)

print("__________________________________________________________")
# we have to calculate future value and total earning
# So for that we will write formulas For

#total Amount Deposited
#Future Value
# Total Earnings

#for monthly interest Rate
i=Expected_RR/(12*100)
#total Number of months in which u invested
N = Inv_duration*12

# now FUTURE VALUE FORMULAA IS

#FV=P[ i(1+i)^n âˆ’1 / i ] (1+i)

# so for future value we need investment duration formula or value and Interest Rate we have both put in formula
#here p=Monthly_Inv_Amo
# n=Inv_duration
Future_Value= Monthly_Inv_Amo*(((1+i)**N -1) / i) *(1+i)
Total_investment= Monthly_Inv_Amo *N #this will tell total investment
Total_Earning=Future_Value-Total_investment

# NOW PRINT THE RESULTS
print("\n***************|  SIP CALCULATOR  |*****************")
print(" -------------------- RESULT ----------------------")

print("\nTHE TOTAL INVESTMENT IS= ",round(Total_investment,2))#round 2 means 2 values after point
print(f"\nAfter {Inv_duration} years, Your Future Investment will grow to = {Future_Value:.2f}") #:.2f this will show only two after points values
print("\nYour Total Earning=",round(Total_Earning,2))
```
