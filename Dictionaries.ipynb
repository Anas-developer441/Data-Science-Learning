```python
# Task 1: The Library Book Tracker(with bonus challenge)

book_data = [
    ["978014", "1984", "George Orwell", 1949],
    ["978020", "To Kill a Mockingbird", "Harper Lee", 1960],
    ["978030", "The Great Gatsby", "F. Scott Fitzgerald", 1925],
    ["978045", "Pride and Prejudice", "Jane Austen", 1813]
]

library_catalog = {}

for row in book_data:
    library_catalog[row[0]] = row[1:]

print("Dictionary is:", library_catalog)

# Access data
print("Title:", library_catalog["978014"][0])

# Membership check
if "978020" in library_catalog:
    print("Found")
else:
    print("Not Found")

# Modify
library_catalog["978014"][2] = 2024

# Remove
library_catalog.pop("978030")

# Bonus add
library_catalog["978040"] = ["The Silent Patient", "Alex Michaelides", 2019]

# Loop print
for isbn, info in library_catalog.items():
    print("Copy of", info[0], "by", info[1], "is available.")







```

```python
#                            Task 2: Employee Shift Manager

#Requirements:

#Create a dictionary named shift_roster.
#Use the employee_data list to populate it. The Key is the Employee ID (Integer),
#and the Value is a dictionary containing {"name": str, "shift": str, "hourly_rate": float}

#Access Data: Retrieve and print the hourly rate of the employee with ID 102.
#Membership Check: Check if Employee ID 105 is in the roster. If they are not, print "Shift coverage needed!"
#Modify Data: Employee 101 got a raise. Update their hourly_rate to 18.50.
#Remove Data: Employee 103 quit today. Remove them from the shift_roster using .pop().
#Bonus Task: Create a function or logic block that checks all employees. If an employee's shift is "Night", add a new key inside their value dictionary called "bonus" with a value of 5.00. Print the updated dictionary for any employee who received this bonus.


# Format: [ID, Name, Shift, Hourly Rate]
employee_data = [
    [101, "Alice Johnson", "Morning", 15.00],
    [102, "Bob Smith", "Afternoon", 16.50],
    [103, "Charlie Brown", "Night", 17.00],
    [104, "Diana Prince", "Morning", 15.50]
]
#Create a dictionary named shift_roster.
shift_roster={}
#Use the employee_data list to populate it.The Key is the Employee ID (Integer)
#and the Value is a dictionary containing {"name": str, "shift": str, "hourly_rate": float}
for i in employee_data:
    shift_roster[i[0]] = {
        "name": i[1],
        "shift": i[2],
        "hourly_rate": i[3]
    }
print("Roster:", shift_roster)

# Accessing the Data
print("Hourly rate of 102 ->", shift_roster[102]["hourly_rate"])

#now  Membership check
if 105 in shift_roster:
    print("Found")
else:
    print("\nShift coverage needed!")

shift_roster[101]["hourly_rate"] = 18.50

# Remove data
shift_roster.pop(103)


# now i am doing Bonus task
for i in shift_roster:

    if shift_roster[i]["shift"] == "Night":

        shift_roster[i]["bonus"] = 5.00

        print("\nBonus added to ->", shift_roster[i])


print("\nFinal roster:", shift_roster)

```

```python
#                       Task 3: Inventory Restock Alert
# Format: [Product Name, Quantity]
stock_data = [
    ["Laptops", 15],
    ["Monitors", 8],
    ["Keyboards", 25],
    ["Mice", 30],
    ["Floppy Disks", 2],
    ["Cpu",2],
]
#initialize inventory dicitonary usong stock_data
inventory={}
for i in stock_data:
    inventory[i[0]] = i[1]  #product names becomes key value

print("Inventory:", inventory)

#Access Data: Print the current quantity of "Laptops".
print("The quantity of laptops only: ",inventory["Laptops"])

if "Tablets" in inventory:
    print("item Tablet in stock ")
else:
    print("Tablet item out of stock")

#Modify/Add Data:
#add "Tablets" with a quantity of 20.
inventory["Tablets"]=20
#increase the quantity of "Monitors" by 5
inventory["Monitors"]+=5
print("\nupdated inventory \n",inventory)

inventory.pop("Floppy Disks")
print("\nInventory without floppy disk after pop\n",inventory)


#                       BONUS TASK
for item in inventory:
    if inventory[item]<10:
        print("\nLow stock of:", item,"Only has",inventory[item],"left\n")
print("\nFINAL INVENTORY-> ",inventory)
```

```python
#                      Extra Task A: Customer Feedback Analyzer

feedback_keywords = [
    "delicious", "slow", "friendly", "delicious", "expensive",
    "slow", "delicious", "clean", "friendly", "slow",
    "expensive", "delicious", "rude", "clean", "delicious"
]


# create an empty dictionary called frequence_table
frequency_table={}

for i in feedback_keywords:

    if i in frequency_table:
        frequency_table[i]+=1
    else:
        frequency_table[i]=1 # this will add in the key if its not in the frequency_table
print(frequency_table)


#Challenge: After building the table, identify and print which keyword appeared the most frequently.

highest_word=""
highest_count=0

for key in frequency_table:
    if frequency_table[key]>highest_count:
        highest_count=frequency_table[key]
        highest_word=key
print("Most frequent Keyword: ",highest_word)
print("count:",highest_count)







```

```python
traffic_logs = [
    "Google", "Facebook", "Direct", "Google", "Twitter",
    "Google", "Direct", "Facebook", "Google", "Instagram",
    "Direct", "Twitter", "Google", "Facebook", "Direct",
    "Google", "LinkedIn", "Direct"
]
source_counts = {}

# frequency table
for i in traffic_logs:

    if i in source_counts:
        source_counts[i] += 1
    else:
        source_counts[i] = 1
print("Source counts:", source_counts)
# total visitors
total_visitors = 0

for key in source_counts:
    total_visitors += source_counts[key]

print("Total visitors:", total_visitors)


# percentage from Google
google_count = source_counts["Google"]

percentage = (google_count / total_visitors) * 100

print("Google traffic % =", percentage)
```
