```python
# i think for better output we can seperate priorites so they dont mix up so what we gonna do is

import csv

# Open file
with open('patient_vitals.csv', encoding='utf-8') as file:
    patientData = list(csv.reader(file))

header = patientData[0]
rows = patientData[1:]

# Column positions
weight_pos = header.index("Weight_kg")
height_pos = header.index("Height_cm")
age_pos = header.index("Age")
heart_pos = header.index("HeartRate")
temp_pos = header.index("Temp_C")
condi_pos = header.index("Condition")

# Separate lists for priorities
critical_list = []
high_list = []
medium_list = []
low_list = []

BMI_list = []

for data in rows:
    weight_kg = float(data[weight_pos])
    height_m = float(data[height_pos]) / 100
    BMI = weight_kg / (height_m ** 2)
    BMI_list.append(BMI)

    age = int(data[age_pos])
    heart = float(data[heart_pos])
    temp = float(data[temp_pos])
    condition = data[condi_pos]

    if condition == "Critical" or (heart > 110 and temp > 39.0):
        priority = "CRITICAL"
        critical_list.append((data[0], round(BMI, 2), priority))
    elif condition == "Observation" and (BMI < 18.5 or BMI > 30.0):
        priority = "HIGH"
        high_list.append((data[0], round(BMI, 2), priority))
    elif age > 75:
        priority = "MEDIUM"
        medium_list.append((data[0], round(BMI, 2), priority))
    else:
        priority = "LOW"
        low_list.append((data[0], round(BMI, 2), priority))

# Print CRITICAL
print("\n--- CRITICAL ---")
for pid, bmi, pri in critical_list:
    print(f"{pid} BMI= {bmi}")
    print("  -> Priority:", pri)

# Print HIGH
print("\n--- HIGH ---")
for pid, bmi, pri in high_list:
    print(f"{pid} BMI= {bmi}")
    print("  -> Priority:", pri)

# Print MEDIUM
print("\n--- MEDIUM ---")
for pid, bmi, pri in medium_list:
    print(f"{pid} BMI= {bmi}")

# Print LOW
print("\n--- LOW ---")
for pid, bmi, pri in low_list:
    print(f"{pid} BMI= {bmi}")

# Average BMI
averageBMI = sum(BMI_list) / len(BMI_list)
print("\nAverage BMI of all patients =", round(averageBMI, 2))

print("\nTotal CRITICAL patients:", len(critical_list))
print("Total HIGH patients:", len(high_list))

# this is a method with seperate listing and printing each list using for loop
```

```python
#                      TASK # 3     WITH BONUS
#                      TASK # 3     WITH BONUS
import csv
file = open('solar_grid.csv',encoding='utf-8')
Energy_Grid=list(csv.reader(file))
Energy_Grid

# seperating header and rows

header=Energy_Grid[0]
rows=Energy_Grid[1:]

#                      index of voltage and current_Amp
V=header.index("Voltage")
CA=header.index("Current_Amps")
EO=header.index("Expected_Output")

W=header.index("Weather")
P=header.index("PanelType")   # or PanelType / PanelID depending on file name

#                   EMPTY LIST TO STORE ACTUAL OUTPUT VALUES
Actual_output=[]


# Efficiency Check: Create a list of differences: Difference = Actual_Output - Expected_Output.
Difference=[]




# bonus part addition
maintenance_actions=[]

replace_count=0
clean_count=0

problem_panels=[]

# NEW LISTS FOR BEAUTIFUL OUTPUT
replace_list=[]
clean_list=[]
check_list=[]
ignore_list=[]
normal_list=[]


#                           COMPUTATION

#convert voltage and current amps to float


for i,row in enumerate(rows,start=1):

    Voltage=float(row[V])
    Current_amp=float(row[CA])
    Expected_Output=int(row[EO])

    weather=row[W]
    panel=row[P]

    #Calculate Actual_Output = Voltage × Current_Amps × 10 (scaling factor).
    Actual_Output=Voltage * Current_amp * 10

    # APPEND Actual_Output values to the list
    Actual_output.append(Actual_Output)

    # Efficiency Checked
    difference = Actual_Output - Expected_Output
    Difference.append(difference)


    # BONUS CONDITIONS

    if weather == "Rainy":
        action = "IGNORE"
        ignore_list.append((panel,i))

    elif Voltage < 22.0:
        action = "CHECK WIRING"
        check_list.append((panel,i))

    elif Actual_Output >= Expected_Output:
        action = "NORMAL"
        normal_list.append((panel,i))

    elif Actual_Output < 0.2 * Expected_Output and weather == "Sunny":
        action = "REPLACE IMMEDIATELY"
        replace_count += 1
        problem_panels.append(panel)
        replace_list.append((panel,i))

    elif Actual_Output < 0.5 * Expected_Output and weather == "Sunny":
        action = "CLEAN PANELS"
        clean_count += 1
        problem_panels.append(panel)
        clean_list.append((panel,i))

    else:
        action = "NORMAL"
        normal_list.append((panel,i))

    maintenance_actions.append(action)







#Calculate Average if Actual Output across All Logs

actual_output_average = sum(Actual_output) / len(Actual_output)


#LIST OF DIFFERENCE ACTUAL OUTPUT- EXPECTED OUTPUT
print("\nLIST OF DIFFERENCE")
for i,data in enumerate(Difference,start=1):
    print(i,round(data,2))



print("\nAverage Actual Output:", actual_output_average)


print("**********************************************")

print("********************   BONUS PART   *************************")


print("\nMaintenance Actions:")
for a in maintenance_actions:
    print(a)


print("\nReplace Immediately:", replace_count)
print("Clean Panels:", clean_count)

print("Problem Panels:", set(problem_panels))


#        tried to  BEAUTIfy OUTPUT

print("\n========== REPLACE IMMEDIATELY ==========")
for p,r in replace_list:
    print("Row",r,"->",p)

print("\n========== CLEAN PANELS ==========")
for p,r in clean_list:
    print("Row",r,"->",p)

print("\n========== CHECK WIRING ==========")
for p,r in check_list:
    print("Row",r,"->",p)

print("\n========== IGNORE ==========")
for p,r in ignore_list:
    print("Row",r,"->",p)

print("\n========== NORMAL ==========")
for p,r in normal_list:
    print("Row",r,"->",p)
```
