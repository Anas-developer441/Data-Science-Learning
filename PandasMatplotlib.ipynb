```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 


df=pd.read_csv('airlines_flights_data.csv')
df.info()



#so in data 11 columns are availale and 30lac+ entries 
# and data is cleansed
```

```python
#                                                         Doing Data inspection 
'''for solving any data analysis project, the first thing a data analyst does is check:

Is the dataset clean?
Are there missing values?
Are column types correct?
Are there duplicate rows?
Are values logical?
Can the data be trusted for analysis?

This process is called Data Cleaning + Data Inspection.'''
```

```python
# check df info and those values are good for calculation in which in front of written that not nulled 
# and bad are those where its written (' object ') this is bad and need to check 

print("\n *******************    INSPECTING FULL DATA     ******************* ")
df.info()

print("\n CHECKING MISSING VALUES")
# check missing values using .isna and sum 
isna=df.isna().sum()
print(isna)

print("\n CHECKING DUPLICATE VALUES ")
#USING DUPLICATE.SUM
duplicate=df.duplicated().sum()
print(duplicate)

describe=df.describe()
print(describe)

# The data is all clean and ready for calculation
```

```python
# TASK 1
# WHAT ARE THE AIRLINES IN THE DATASET, ACCOMPANIED BY THEIR FREIQUENCIES(like how many flights are present in the market )

AirLine_counts=df['airline'].value_counts()
print(AirLine_counts)

# visualiation
#using plt
AirLine_counts.plot(kind='bar')
plt.title('Airlines and their Frequencies')
plt.xlabel('AirLines')
plt.ylabel('Frequency')

plt.xticks(rotation=45)
plt.show()
```

```python
#Q2. Show Bar Graphs representing the Departure Time & Arrival Time
dpt_counts=df['departure_time'].value_counts()
print(dpt_counts)

dpt_counts.plot(kind='bar')
plt.title('Departure time Distribution')
plt.xlabel('Departure Time')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
print("**********************************************************888")

arr_count=df['arrival_time'].value_counts()
print(arr_count)

arr_count.plot(kind='bar')
plt.title('Arrival Time Distribution')
plt.xlabel('Arrival Time')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q3. Show Bar Graphs representing the Source City & Destination City
source_counts = df['source_city'].value_counts()

print(source_counts)
source_counts.plot(kind='bar')

plt.title('Flights from Source Cities')
plt.xlabel('Source City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()




print("***************************************")
destination_counts = df['destination_city'].value_counts()

print(destination_counts)
destination_counts.plot(kind='bar')

plt.title('Flights to Destination Cities')
plt.xlabel('Destination City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q4. Does price vary with airlines?

avg_price_airline=df.groupby('airline')['price'].mean()
print(avg_price_airline)


avg_price_airline.plot(kind='bar')

plt.title('Average Ticket Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

```

```python
#Q5. Does ticket price change based on departure time and arrival time?
tckt_price=df.groupby('departure_time')['price'].mean()
print(tckt_price)

tckt_price.plot(kind='bar')

plt.title('Average Price by Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()


print("*******************************8")
arr_price = df.groupby('arrival_time')['price'].mean()

print(arr_price)
arr_price.plot(kind='bar')

plt.title('Average Price by Arrival Time')
plt.xlabel('Arrival Time')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q6. How does the price change with change in Source and Destination?
source_price = df.groupby('source_city')['price'].mean()

print(source_price)
source_price.plot(kind='bar')

plt.title('Average Price by Source City')
plt.xlabel('Source City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

destination_price = df.groupby('destination_city')['price'].mean()

print(destination_price)
destination_price.plot(kind='bar')

plt.title('Average Price by Destination City')
plt.xlabel('Destination City')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#Q7. How is the price affected when tickets are bought in just 1 or 2 days before departure?
df.head()
booking=df[df['days_left']<=2]
print(booking[['days_left','price']].head())

avg=booking.groupby('days_left')['price'].mean()
print("\n",avg.head())

avg.plot(kind='bar')

plt.title('Average Price for Last Minute Booking')
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q8. How does the ticket price vary between Economy and Business class?

ticket_price=df.groupby('class')['price'].mean()
print(ticket_price)

ticket_price.plot(kind='bar')

plt.title('Average Ticket Price by Class')
plt.xlabel('Class')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q9. What will be the Average Price of Vistara airline for a flight from Delhi to Hyderabad in Business Class?
vistara_flight = df[
    (df['airline'] == 'Vistara') &
    (df['source_city'] == 'Delhi') &
    (df['destination_city'] == 'Hyderabad') &
    (df['class'] == 'Business')
]

average_price = vistara_flight['price'].mean()

print("Average Price:", average_price)

```
