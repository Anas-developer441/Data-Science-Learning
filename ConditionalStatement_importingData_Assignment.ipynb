```python
# i think for better output we can seperate priorites so they dont mix up so what we gonna do is

import csv

# Open file
with open('patient_vitals.csv', encoding='utf-8') as file:
    patientData = list(csv.reader(file))

header = patientData[0]
rows = patientData[1:]

# Column positions
weight_pos = header.index("Weight_kg")
height_pos = header.index("Height_cm")
age_pos = header.index("Age")
heart_pos = header.index("HeartRate")
temp_pos = header.index("Temp_C")
condi_pos = header.index("Condition")

# Separate lists for priorities
critical_list = []
high_list = []
medium_list = []
low_list = []

BMI_list = []

for data in rows:
    weight_kg = float(data[weight_pos])
    height_m = float(data[height_pos]) / 100
    BMI = weight_kg / (height_m ** 2)
    BMI_list.append(BMI)

    age = int(data[age_pos])
    heart = float(data[heart_pos])
    temp = float(data[temp_pos])
    condition = data[condi_pos]

    if condition == "Critical" or (heart > 110 and temp > 39.0):
        priority = "CRITICAL"
        critical_list.append((data[0], round(BMI, 2), priority))
    elif condition == "Observation" and (BMI < 18.5 or BMI > 30.0):
        priority = "HIGH"
        high_list.append((data[0], round(BMI, 2), priority))
    elif age > 75:
        priority = "MEDIUM"
        medium_list.append((data[0], round(BMI, 2), priority))
    else:
        priority = "LOW"
        low_list.append((data[0], round(BMI, 2), priority))

# Print CRITICAL
print("\n--- CRITICAL ---")
for pid, bmi, pri in critical_list:
    print(f"{pid} BMI= {bmi}")
    print("  -> Priority:", pri)

# Print HIGH
print("\n--- HIGH ---")
for pid, bmi, pri in high_list:
    print(f"{pid} BMI= {bmi}")
    print("  -> Priority:", pri)

# Print MEDIUM
print("\n--- MEDIUM ---")
for pid, bmi, pri in medium_list:
    print(f"{pid} BMI= {bmi}")

# Print LOW
print("\n--- LOW ---")
for pid, bmi, pri in low_list:
    print(f"{pid} BMI= {bmi}")

# Average BMI
averageBMI = sum(BMI_list) / len(BMI_list)
print("\nAverage BMI of all patients =", round(averageBMI, 2))

print("\nTotal CRITICAL patients:", len(critical_list))
print("Total HIGH patients:", len(high_list))

# this is a method with seperate listing and printing each list using for loop
```
