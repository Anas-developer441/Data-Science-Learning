```python
import pandas as pd

# 1. Load the data fresh
data = pd.read_csv('googleplaystore.csv')

# 2. Clean only if it's text.
# We use .str.replace on the 'Installs' column (with an 's')
data['Installs'] = data['Installs'].str.replace('+', '').str.replace(',', '')

# 3. Convert to numeric (Fixed the typo from 'Install' to 'Installs')
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# 4. Drop the NaN row and duplicates
data = data.dropna(subset=['Installs'])
data_unique = data.drop_duplicates(subset='App')

# 5. Get Top 5 (Fixed the typo 'nargest' and the 'unique' variable name)
top_5_apps = data_unique.nlargest(5, 'Installs')

# Display the result
print(top_5_apps[['App', 'Installs']])
```
